## Методы решения систем линейных уравнений
СЛАУ - система уравнений, каждое уравнение в которой является линейным - алгебраическим уравнением первой степени.



### Метод Гаусса
Это метод последовательного *исключения* переменных, когда с помощью элементарных преобразований система уравнений *приводится к равносильной системе* правого верхнего (традиционного), правого нижнего, левого верхнего или левого нижнего треугольного вида, из которой последовательно, начиная с последних или c первых (по номеру), находятся все переменные системы.

СЛАУ



1) Записать расширенную матрицу системы



В расширенной матрице добавляется столбец свободных членов.

2) Произвести элементарные преобразования:
- Строки можно переставлять местами



- Если в матрице есть пропорциональные или одинаковые строки, то следует удалить из матрицы все такие строки кроме одной.



- Если в матрице появилась нулевая строка, ее нужно удалить.

- Строку матрицы можно умножить/разделить на любое число, отличное от 0 (Здесь первую строку разделили на -3, а вторую умножили на 2).
