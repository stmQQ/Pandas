## Работа с матрицами
Матрица - математический объект, записываемый в виде прямоугольной таблицы. Количество строк и столбцов задает размер матрицы. 

| | 0 | 1 | 2 |
|-|-|-|-|
|0|a<sub>00</sub>|a<sub>01</sub>|a<sub>02</sub>|
|1|a<sub>10</sub>|a<sub>11</sub>|a<sub>12</sub>|
|2|a<sub>20</sub>|a<sub>21</sub>|a<sub>22</sub>|

Матрицы естественным образом возникают при решении систем линейных уравнений.
Рассмотрим систему линейных уравнений:



Эта система состоит из *m* линейных уравнений относительно *n* неизвестных. ЕЕ можно записать в виде:

**Ax = b**, где



Если у матрицы количество строк *m* совпадает с количеством столбцов *n*, то такая матрица называется **квадратной**, а число *m* называется **порядком** матрицы.

#### Транспонированная матрица
Для каждой матрицы A можно построить матрицу B, у которой b<sub>ji</sub> = a<sub>ij</sub> для всех i и j.

Такая матрица называется транспонированной для A и обозначается A<sup>T</sup>.
При транспонировании матрицы строки становятся столбцами, а столбцы - строками.



#### Единичная матрица
Единичная матрица - квадратная матрица E, элементы на главной диагонали которой равны единице (e<sub>ii</sub> = 1), а остальные равны нулю (e<sub>ij</sub> = 0 при i &#8800; j).



#### Нулевая матрица
Нулевая матрица - матрица, все элементы которой равны 0.



### Операции над матрицами
#### Сложение матриц
Складывать можно только матрицы одинакового размера.
Сложение матриц A + B - операция нахождения матрицы C, где 

**c<sub>ij</sub> = a<sub>ij</sub> + b<sub>ij</sub>**.

#### Умножение матрицы на число
Умножение матрицы A на число k заключается в построении матрицы B, где 

**b<sub>ij</sub> = (k * a<sub>ij</sub>)**

#### Умножение матриц
Умножение матриц A &times; B есть операция вычисления матрицы C, каждый элемент которой равен сумме произведений элементов в соответствующей строке первого множителя и столбце второго:

**с<sub>ij</sub> = &Sigma;(a<sub>ik</sub>b<sub>kj</sub>)**

Матрица A должна быть согласованной с матрицей B (количество столбцов A должно быть равно количеству строк B). Если матрица A имеет размер *m* &times; *n*, B - *n* &times; *k*, то размерность матрицы C - *m* &times; *k*.
