## Распределения
Распределение - это закон, который описывает, как вероятности распределены между всеми возможными значениями случайной величины.

### Нормальное распределение
Нормальное распределение (распределение Гаусса, колоколообразное распределение) — это тип распределения, при котором большинство значений сосредоточено вокруг среднего. График — симметричная колоколообразная кривая.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# loc - среднее, scale - стандартное отклонение, size - размер выборки
data_normal = np.random.normal(loc=50, scale=10, size=1000)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data_normal, bins=30, density=True, alpha=0.7, color='skyblue', 
         edgecolor='black', label='Данные')

# Теоретическая кривая Гаусса
x = np.linspace(20, 80, 1000)  # Диапазон значений x
pdf = norm.pdf(x, loc=50, scale=10)  # Функция плотности вероятности
plt.plot(x, pdf, 'r-', linewidth=2, label='Теоретическая кривая N(50, 10)')

# Настройки графика
plt.title('Нормальное распределение: данные и теоретическая кривая')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

- Среднее значение **μ** (мю) и стандартное отклонение **σ** (сигма) определяют форму кривой нормального распределения (гауссианы).
- По закону нормального распределения 68,2% значений находятся в пределах одного σ от μ, 95,4% — в пределах двух σ, а 99,7% — в пределах трёх σ.
- Зная среднее значение и стандартное отклонение распределения, можно устанавливать контрольные границы и принимать решения на их основе.

### Биноминальное распределение
Биноминальное распределение - это распределение вероятностей, которое описывает количество успехов в фиксированном количестве независимых испытаний, где каждое испытание имеет только два возможных исхода: успех или неудача.

### Распределение Бернулли
Распределение Бернулли - частный случай биноминального распределения, где количество испытаний **n** равно 1.

P(X=1) = p (вероятность успеха)

P(X=0) = 1-p (вероятность неудачи)

E[X] = p (математическое ожидание)

Var[X] = p(1-p) (дисперсия)


```python
from scipy.stats import bernoulli

p = 0.3
data = bernoulli.rvs(p, size=10)
print(data)

print(f"P(X=1) = {bernoulli.pmf(1, p):.2f}")
print(f"P(X=0) = {bernoulli.pmf(0, p):.2f}")
```
